# TODO:

In the code below, write a function `get_num_components(df, var_ret)` that takes a dataframe, `df`, and a value for the desired amount of variance you want to retain from the `df` dataframe,`var_ret`. In this case, the parameter `df` should be the `returns` dataframe obtained above. The parameter  `var_ret` must be a number between 0 and 1. The function should return the number of principal components you need to retain that amount of variance. To do this, use Scikit-Learn's PCA() class and its `.explained_variance_ratio_`. The function should also print the total amount of variance retained. 

```python
# import resources
from sklearn.decomposition import PCA
import numpy as np

# I need to store a tuple current step, previous ste

def get_num_components(df, var_ret):
    
    # initialisation
    
    variance_expl = [1, 1]     # list recording current and previous value of variance explained
    step = 50                  # steps to take out at each iteration
    tot_dim = df.shape[1]      # total number of dimensions
    dim_ret = df.shape[1] - 1  # starting value for dimension - hyp all the dimensions explain all the variance
    it = 1                     # iteration count
    
    while dim_ret > 1:
        
        # execute the PCA and calculate the variance explained
        pca = PCA(n_components = dim_ret)        
        pca.fit(df)                
        curr_explained_var = np.sum(pca.explained_variance_ratio_)
        tmp = variance_expl[0]
        variance_expl[0] = curr_explained_var         
                
        if variance_expl[0] > var_ret and variance_expl[1] > var_ret:            
                                    
            # update the record of the old variance explained
            variance_expl[1] = tmp
        
        elif (variance_expl[0] < var_ret and variance_expl[1] > var_ret and step > 1):            
            dim_ret += step - 1
            step = int( np.floor(step / 2) )
            
            # update the record 
            pca = PCA(n_components = dim_ret+1)
            pca.fit(df)                
            curr_explained_var = np.sum(pca.explained_variance_ratio_)            
            variance_expl[1] = curr_explained_var
            
            print(f"Component moved back to {dim_ret} and step halved. Current step {step}")
                        
        elif (variance_expl[0] < var_ret and variance_expl[1] > var_ret and step == 1):
            dim_ret += step + 1       
            
        else:
            if step > 1:
                dim_ret += step - 1
                step = int( np.floor(step / 2) )
                
                # update the record 
                pca = PCA(n_components = dim_ret+1)
                pca.fit(df)                
                curr_explained_var = np.sum(pca.explained_variance_ratio_)            
                variance_expl[1] = curr_explained_var
            else:
                break
                
        print(f"# {it} {dim_ret} components explain {variance_expl[0]:.6f} - prev. var {variance_expl[1]:.6f}")        
        
        if dim_ret - step > 0:
            dim_ret -= step
        else:
            dim_ret -= 1
            
        it += 1
        
    return dim_ret+1     
            
num_components = get_num_components(returns, 0.5)

print('\nNumber of Principal Components Needed: ', num_components)
```

```python
# To verify
pca = PCA(n_components = 7)
pca.fit(returns)        
curr_explained_var = np.sum(pca.explained_variance_ratio_)
print(f"explained variance {curr_explained_var:.6f}")
```
